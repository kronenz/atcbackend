```python

from ncclient import manager
from ncclient.xml_ import to_ele

klm_host = "10.0.20.150"
klm_user = "admin"
klm_pwd = "Forwiz_SDF1"


# Connect to the Kaloom Netconf server

cnn = manager.connect(
    host=klm_host,
    port=830,
    username=klm_user,
    password=klm_pwd,
    hostkey_verify=False)
    # Build the RPC request using the Kaloom Netconf data model

```

```python
rpc_request = to_ele("""<get><filter type="subtree"><data xmlns:fabrics="urn:kaloom:faas:fabrics"></data>
                        <Fabric></Fabric>
                     </filter></get>""")

print(rpc_request)
try:
    response = cnn.dispatch(rpc_request).xml
    print(response)
except Exception as e:
    print("error")
    print(str(e))
```
